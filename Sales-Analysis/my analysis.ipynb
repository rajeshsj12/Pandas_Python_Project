### 16.	Total sales values by year and month.

```python
df['order_month'] = df['order_date'].dt.month_name()
df['order_monthno'] = df['order_date'].dt.month
df['mon_year'] = df['year'].astype(str) +" "+ df['order_month']
```

```python
sales_by_year_month = df.groupby(['year','order_month','order_monthno','mon_year'],as_index=False).aggregate({'sales':'sum'}).sort_values(by=['year','order_monthno'])
sales_by_year_month.reset_index(drop=True,inplace=True)
sales_by_year_month
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_style('whitegrid')
fig, axs = plt.subplots(2, figsize=(15, 10))

# Create the barplot on axs[0]
sns.barplot(data=sales_by_year_month, x='year', y='sales', hue='order_month', palette='tab20', ax=axs[0])
axs[0].set(xlabel='Year Month', ylabel='Sales', title='Sales by Month')
axs[0].legend(title='Sub-Category', loc='upper left', bbox_to_anchor=(1, 1))

# Add data labels to the bars on axs[0]
for p in axs[0].patches:
    axs[0].annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', fontsize=10, color='white', backgroundcolor='blue', xytext=(0, 5),
                    textcoords='offset points')

# Create the lineplot on axs[1]
sns.lineplot(data=sales_by_year_month, x='mon_year', y='sales', ax=axs[1])
axs[1].set_xticklabels(axs[1].get_xticklabels(), rotation=90)
axs[1].set(xlabel='Month Year', ylabel='Sales', title='Line Graph')

# Set the figure title
fig.suptitle('Total sales values by year and month', fontsize=16, y=1.05, fontweight = 'bold')

fig.tight_layout()
plt.show()
```

```python
sns.set_style('whitegrid')
fig, axs = plt.subplots(figsize=(10, 5), dpi=500)

# Create the lineplot on axs
sns.lineplot(data=sales_by_year_month, x='mon_year', y='sales', ax=axs)
axs.set_xticklabels(axs.get_xticklabels(), rotation=90)
axs.set(xlabel='Month Year', ylabel='Sales', title='Line Graph')

# Calculate and plot the SMA line
sales_by_year_month['sma'] = sales_by_year_month['sales'].rolling(window=3).mean()
sns.lineplot(data=sales_by_year_month, x='mon_year', y='sma', color='red', ax=axs, label='SMA (3-Month)')

# Set the figure title
fig.suptitle('Total sales values by year and month', fontsize=16, y=1.05, fontweight='bold')

fig.tight_layout()
plt.show()
```
